```python
def lcg(M, a, b, N0, k):
    """"
    Function that implements a linear congruential generator
    
    Input: 
          M     -> Given int
          a     -> Given int
          b     -> Given int
          N0    -> Seed
          k     -> Dimension
          
    Output:
         U      -> Pseudo random numbers
    """
    U = np.zeros(k)
    U[0] = N0 
    
    for i in range(1 , k):
        U[i] = (a*U[i-1] + b) % M

    

    return U/M
```

```python
def m_polar(U1, U2, k):
    """
    Function that implements Marsaglia's polar algorithm
    
    Input:
        U1 -> Uniform deviates 
        U2 -> Uniform deviates
        k  -> length of above RV
    
    Output:
        Z1 -> Normal deviates
        Z2 -> Normal deviates
        k_new -> length of above RV
    """
    Z1 = []
    Z2 = []
    # S = [u1**2 + u2**2 for u1 , u2 in zip(U_100[f"k = {k}"] , U_1[f"k = {k}"]) if u1**2 + u2**2 <= 4 and u1**2 + u2**2 >= -4]

    for i in range(k):
        u1 = 2*U1[i] -1
        u2 = 2*U2[i] -1
        S = u1**2 + u2**2
        
        if S >= 1 or S == 0 :
            continue
        factor = np.sqrt(-2 * np.log(S) / S)
        z1 = u1*factor
        z2 = u2*factor
        if -4 <= z1 <= 4 and -4 <= z2 <= 4:
            Z1.append(z1)
            Z2.append(z2)
    
    return Z1 , Z2 , len(Z1)
    # Add code here
```
